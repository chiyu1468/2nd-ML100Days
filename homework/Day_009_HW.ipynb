```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
get_ipython().magic('matplotlib inline')
```

 ### 負相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
# noise 越高, 相關度越低
noise = 20
x = np.random.randint(0, 50, 1000)
y = 25 - x + np.random.normal(0, noise, 1000)
```

```python
# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x, y)
```

```python
# 將分布畫出來看看吧
plt.scatter(x, y)


```
